```python
def parse_kitti_label(kitti_label_file):
    """
    Parse KITTI format label file. KITTI format is a filename corresponding to the image filename with a line of
    annotations for each instance of an object in the format <class_name, 0.00, 0, 0.00, x_min, y_min, x_max, y_max>
    :param kitti_label_file: Filename of KITTI file to parse
    :return: [<list_of_bboxes in xmin, ymin, xmax, ymax format>], [<list_of_class_ids>]
    """

    # Parse file and get labels and bbox coordinates
    bboxes = []
    labels = []
    for line in open(kitti_label_file).readlines():
        line_contents = line.split(' ')
        assert len(line_contents) >= 8, f'File {kitti_label_file} does not have required number of fields'
        labels.append(line_contents[0])
        bbox = [int(float(line_contents[4])), int(float(line_contents[5])),
                int(float(line_contents[6])), int(float(line_contents[7]))]
        bboxes.append(bbox)

    return bboxes, labels
```
